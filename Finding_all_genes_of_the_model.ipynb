```python
import numpy as np
import pandas as pd
import cobra
import os
from os.path import join

from time import time

a= cobra.io.read_sbml_model("iEC1344_C.xml")
```

```python
cobra_config = cobra.Configuration()
cobra_config.solver = "gurobi"
cobra_config.solver
```

```python
a.genes[0].name
```

```python
i= 0
model_gene_names= list()

while i < len(a.genes):
    model_gene_names.append(a.genes[i].name)

    i= i+1

model_gene_names
```

```python
import pandas as pd

d= pd.DataFrame(model_gene_names, columns= ["Gene names for all reactions"])
d.to_csv("all genes.csv")
```
